#### Observations

There are no high correlations between Compressive strength and other variables except for Cement. Cement has a 0.5 correlation coefficient, meaning that cement and concrete strength are directly proportional.

Age and Super plasticizer are the other two variables strongly correlated with Compressive Strength.

Super Plasticizer seems to have a high negative correlations with Water and positive correlations with Fly ash and Fine aggregate

Age has a relatively strong correlation with water as well.

Cement, Age, Superplasticizer, Water have significant correlations with Concrete Strength



Due to multi colinearity we cannot use Superplaticizer and water in the model.

Therefore we will continue to make our model taking Cement, Age, Superplasticizer as the independant variables

## Multi-linear model

```python
#split dataset in features and target variable
x = concrete_data.iloc[:, [0,4,7]]         # Features - Cement, Age, Superplasticizer
y = concrete_data.iloc[:,[8]]          # Target - CC_Strength
```

```python
np.random.seed(123)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)

print(f'alpha = {model.intercept_}')
print(f'betas = {model.coef_}')
```

```python
#Testing the model on the training set
y_train_pred = model.predict(x_train)

plt.scatter(y_train,y_train_pred)
plt.plot([y_train.min(), y_train.max()], [y_train_pred.min(), y_train_pred.max()], 'r', lw=2)
plt.xlabel("Y_train")
plt.ylabel("Y_train_pred")
plt.title("Predictions vs. actual values in the training set")
```

```python
#Testing the model on the test set
y_test_pred = model.predict(x_test)

plt.scatter(y_test,y_test_pred)
plt.plot([y_test.min(), y_test.max()], [y_test_pred.min(), y_test_pred.max()], 'r', lw=2)
plt.xlabel("Y_test")
plt.ylabel("Y_test_pred")
plt.title("Predictions vs. actual values in the test set")
```

```python
#Finding the MSE and the RMSE values for the predictions made on the training set
from sklearn import metrics
RMSE_train = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))
print("The RMSE of the predictions made on training set: ",round(RMSE_train,3))

#Computing the MSE and the RMSE values for the predictions made on the test set
RMSE_test = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
print("The RMSE of the predictions made on test set: ",round(RMSE_test,3))
```

```python
# copy of the concrete data without the concrete strength 
copy = concrete_data.copy()
```

```python
#The model as a function
def multi_model(df):
    x = df.iloc[:, [0,4,7]]   
    y = df.iloc[:,[8]]
    np.random.seed(123)

    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
    
    from sklearn.linear_model import LinearRegression
    model = LinearRegression()
    model.fit(x_train,y_train)
    
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    from sklearn import metrics
    RMSE_train = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))
    print("The RMSE of the predictions made on training set: ",round(RMSE_train,3))

    RMSE_test = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
    print("The RMSE of the predictions made on test set: ",round(RMSE_test,3))
```

```python
# function to add data to a copy of the concrete data

def predict():
    global copy
    prompt = ["Cement (component 1)(kg in a m^3 mixture): ",
              "Blast Furnace Slag (component 2)(kg in a m^3 mixture): ",
              "Fly Ash (component 3)(kg in a m^3 mixture): ",
              "Water (component 4)(kg in a m^3 mixture): ",
              "Superplasticizer (component 5)(kg in a m^3 mixture): ",
              "Coarse Aggregate (component 6)(kg in a m^3 mixture): ",
              "Fine Aggregate (component 7)(kg in a m^3 mixture): ",
              "Age (day): "]
    vals = []

    print("Input values for the respective data")
    for i in range(0, len(prompt)):
        ans = float(input(prompt[i]))
        vals.append(ans)

    y_pred = float(model.predict([[vals[0], vals[4], vals[7]]]))
    vals.append(y_pred)
    print("The estimated Concrete Strength for your data is ", y_pred)
    copy.loc[len(copy)] = vals
    multi_model(copy)
#     return y_pred

    add = input("Do you want to save this data for your model?(Y/N): ")
    add = add.upper()
    if add == "Y":
        return("Done")
    elif add == "N":
        copy.drop(labels=len(copy)-1, axis=0)
        return("Done")
    else:
        copy.drop(labels=len(copy)-1, axis=0)
        return("Done")
```
